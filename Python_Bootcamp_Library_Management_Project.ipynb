```python
menu_string = "\n*** MENU*** \n 1) List Books \n 2) Add Book \n 3) Remove Book \n 4) Quit \n 5) Find Book \n 6) Search by year \n 7) Search by author \n Enter your choice (1-7):"

while 1:
  lib=Library("books.txt")
  try:
    selection=int(input(menu_string))

    if selection==1:
      lib.list()
      del lib

    elif selection==2:
      lib.add()
      del lib

    elif selection==3:
      lib.remove()
      del lib

    elif selection==4:
      print("Program exited.")
      break

    elif selection==5:
      lib.find()
      del lib

    elif selection==6:
      lib.search_by_year()
      del lib

    elif selection==7:
      lib.search_by_author()
      del lib




  except:
    print("Lütfen geçerli bir numara giriniz. (1-7)")


```

```python
class Library:

  def __init__(self,file_name):
    self.file=open(file_name,"a+")


  def __del__(self):
    self.file.close()


  def list(self):

    self.file.seek(0)
    content=self.file.read()
    list=content.splitlines()



    if len(list)==0:
      print("No books found.")
    else:
       for i in content.splitlines():
         book=i.split(",")
         print(f"Book Title:{book[0]},Book Author:{book[1]}")


  def add(self):



    while True:
       book_title=input("Book Title:")
       if book_title:
        break
       print("Book title cannot be empty.")

    while True:
       book_author=input("Book Author:")
       if book_author:
        break
       print("Book author cannot be empty.")

    while True:
       first_release_year=input("First Release Year:")
       if first_release_year.isdigit():
        break
       print("Please enter numbers only.")
    while True:

       pages_number=input("Number of Pages:")
       if pages_number.isdigit():
        break
       print("Please enter numbers only.")


    info=f"{book_title},{book_author},{first_release_year},{pages_number}\n"

    try:
      self.file.write(info)
      print("Book added.")
    except:
      print("The book could not be added.")

  def find(self):


    book_title=input("Enter the book you want to find:")

    count=False

    self.file.seek(0)
    content=self.file.read()

    for i in content.splitlines():
      book=i.split(",")

      if book[0]==book_title:
        count=True
        print(f"Book Title:{book[0]},Book Author:{book[1]},First Release Year:{book[2]},Number of Pages:{book[3]}")
        break

    if not count:
      print("Book not found.")

  def remove(self):

    book_title=input("Enter the book title to remove:")

    is_find=False
    count=0

    self.file.seek(0)
    content=self.file.read()

    line_list=content.splitlines()

    for index in range(len(line_list)):
      value = line_list[index]
      book=value.split(",")
      if book[0]==book_title:
        count=index
        is_find=True
        break


    if not is_find:
      print("Book not found.")
    else:
      del line_list[count]
      self.file.truncate(0)
      for i in line_list:
        self.file.write(i+"\n")
      print("Book removed.")

  def search_by_year(self):

    year=input("Enter the year you want to find:")

    is_find=False

    self.file.seek(0)
    content=self.file.read()

    for i in content.splitlines():
      book=i.split(",")
      if book[2]==year:
        is_find=True
        print(f"Book Title:{book[0]},Book Author:{book[1]},First Release Year:{book[2]},Number of Pages:{book[3]}")


    if is_find:
      return
    else:
      print("No books found.")



  def search_by_author(self):

    author=input("Enter the author you want to find:")

    empty_list=[]
    self.file.seek(0)
    content=self.file.read()

    is_find=False

    self.file.seek(0)
    content=self.file.read()

    for i in content.splitlines():
      book=i.split(",")
      if book[1]==author:
        is_find=True
        print(f"Book Title:{book[0]},Book Author:{book[1]},First Release Year:{book[2]},Number of Pages:{book[3]}")


    if is_find:
      return
    else:
      print("No books found.")
```
